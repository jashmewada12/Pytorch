#### To reduce the randomnees in neural networks and torch we use seeds
#### It adds flavour to randomness

```python
rand_tensor_1 = torch.rand(9,9)
rand_tensor_2 = torch.rand(9,9)
print(rand_tensor_1)
print(rand_tensor_2)
print(rand_tensor_2 == rand_tensor_1)
```

## making random but reproduccable tensors 
* Note : use torch.manual_seed everytime while creating a tensor to create same random tensors

```python
import torch
seed = 67

torch.manual_seed(seed)#for  random tensor 3
random_tensor_3 = torch.rand(9,9)
torch.manual_seed(seed)# for random tensor 4 
random_tensor_4 = torch.rand(9,9)

print(random_tensor_3)
print(random_tensor_4)
print(random_tensor_3 == random_tensor_4)
```

#### Command to check if pytorch can access the gpu

```python
import torch
torch.cuda.is_available()
```

```python
#Setup device agonstic code
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
#count the number of devies 
torch.cuda.device_count()
```

```python
#using the cpu
tensor = torch.rand(9,9)
print(tensor,tensor.device)
```

```python
#Moving tenosr to gpu
tensor_gpu = tensor.to(device)
tensor_gpu
```

```python
#if tensor is on gpu you cannot transform it into numpy
tensor_gpu.numpy()
```

```python
# fixing the error by transfering the tensor to cpu and then converting to numpy 
tensor_cpu = tensor_gpu.cpu().numpy()
tensor_cpu
```

```python
# the tensor on the gpu remains unchanged when we transfer it back to the cpu
tensor_gpu
```
